```python
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize, sent_tokenize
import cltk
import pandas as pd
import numpy as np
import os
```

## Formatting text (XML to Tuple of desired attributes)

```python
def desoupify(fileloc):
    #Creates soup object
    with open( fileloc, 'r') as f:
        file = f.read() 
    xml_soup = BeautifulSoup(file,'xml')
    
#grabs text from the object 
    text_soup = xml_soup.find_all("p")
   
    length = len(text_soup)
    if length == 0:
        text_soup = xml_soup.find_all("l")

#joins into text
    text = []
    for x in text_soup: 
        text.append(x.text)
    text = ' '.join(text)
    return text
```

```python
def text_clear(text):

#converts all characters to lower case 
    string = text.lower()

    string = string.replace("\n"," ")

#cleans out punctuation
    punc = '''()-[]{};:'"\,<>/@,#$%^&,:*_~'''
    for sym in string:
        if sym in punc:
            string = string.replace(sym,"")
    string = string.replace('keyboarding','')
    return string
```

```python
#tokenizes
def tokenizer_pipline(string): 
    finished = []
    string = sent_tokenize(string)
    for x in string: 
        finished.append(word_tokenize(x))
        
    return finished
    
    
```

```python
#Creates tuples of desired atrributes
def attributor(directory,period): 
    final = []
    
    for filename in os.listdir(directory)[1:]:
        f = os.path.join(directory, filename)
        ready = tokenizer_pipline(text_clear(desoupify(f)))
        for x in ready: 
            y =' '.join(x)
            final.append((directory,period,filename,x,y))
    return final
    
```

## Tuples to Dataframe

```python
x = attributor('Virgil','Golden')
```

```python
data = x

# Creating a DataFrame
df = pd.DataFrame(data,columns=['Author','Period','Work','Tokenized Text', 'Untokenized Text'])

df
```
